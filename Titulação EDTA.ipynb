```python
import matplotlib.pyplot as plt
import numpy as np
```

Considere a titulação de 50,0 mL de Ca+2 0,00500 mol/L  com EDTA 0,0100 mol/L em uma solução tamponada a
pH constante igual a 10,0.
$
\
a) Antes \ da \ adição \ de \ titulante \\
b) após \ a \ adição \  de \  20 mL \ de \ titulante \\
c) No \ ponto \ de \ equivalência \\
d) Após \ aadição \ de \ 60 mL \ de \ titulante \\
\
$

```python
#n = [C0_Ca,V0_Ca,C_EDTA]
n = [0.005,50,0.01]
Volume =np.linspace(0,60,1000) # mL
```

```python
def calculo_tit(n,Volume,Kabs = None, alpha = None):
    
    'log de Kabs e o alpha são tabelado'
    
    #Calculando a constante Condicional (K_linha)
    K_linha = Kabs*alpha
    
    #Calculando Ve :
    C0_Ca = n[0]
    V0_Ca = n[1]
    C_EDTA = n[2]
    Ve = C0_Ca*V0_Ca/C_EDTA
    
    #Titulando 
    
    concentracao = []
    p_Ca = []
    for index, item in enumerate (Volume):
        
        if item<Ve:
            Con_Ca = (C0_Ca*V0_Ca - C_EDTA*Volume[index])/(V0_Ca + Volume[index])
            pCa = -np.log10(Con_Ca)
            concentracao.append(Con_Ca) , p_Ca.append(pCa)
        
        elif item == Ve:
            Con_Complexo = C_EDTA*Ve/(Ve+V0_Ca)
            Con_Ca = np.sqrt(Con_Complexo/K_linha)
            pCa = -np.log10(Con_Ca)
            concentracao.append(Con_Ca) , p_Ca.append(pCa)
            
        else:
            Con_analitica = (C_EDTA*Volume[index] - C0_Ca*V0_Ca)/(V0_Ca + Volume[index])
            Con_Complexo = (C0_Ca*V0_Ca)/(Volume[index] + V0_Ca)
            Con_Ca = Con_Complexo/(K_linha*Con_analitica)
            pCa = - np.log10(Con_Ca)
            concentracao.append(Con_Ca) , p_Ca.append(pCa)
           
    
    display(f"O volume de equivalência é {Ve} mL e o K' é igual a {K_linha}") 
    return concentracao , p_Ca
     

            
            
            
```

```python
#Questao 1
x = calculo_tit(n,Volume,Kabs = 5*10**10, alpha = 0.35)[0] # Concentração do cátion
y = calculo_tit(n,Volume,Kabs = 5*10**10, alpha = 0.35)[1] # Equivalente a pH
calculo_tit(n,Volume,Kabs = 5*10**10, alpha = 0.35)          
```

```python
plt.style.use('tableau-colorblind10')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(Volume,y,label = 'Curva de Titulação')
ax.legend(loc = 'lower right')
plt.xlabel('Volume(mL)')
plt.ylabel('pCa')
plt.title('Curva de titulação')
plt.grid()
plt.ylim(0,14)
plt.xlim(0,60)
fig.tight_layout()
plt.savefig('Curva de titulação Ca+2',dpi = 300)
```
