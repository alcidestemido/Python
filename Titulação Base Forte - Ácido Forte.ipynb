```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
```

```python
#    Acido forte Base forte
'n = [C_basei,Volume_basei,C_acid]'
n = (0.01 ,50 ,0.1)
Volume = (0,1,2,3,4,4.5,4.9,4.99,5,5.01,5.1,6.0,8.0,10)
```

```python
def Calculo_Titulacao(n,Volume):
    C_OHi = n[0]
    Vol_OHi = n[1]
    C_H = n[2]
    #Calculando o Ve
    Ve = C_OHi*Vol_OHi/C_H
    p_H = []
    
    for index,item in enumerate(Volume):
        if item < Ve:
            C_OH = (C_OHi*Vol_OHi - C_H*Volume[index])/(Vol_OHi + Volume[index])
            pH = 14 + np.log10(C_OH)
            p_H.append(pH)       
        elif item == Ve: 
            pH = 7
            p_H.append(pH)
        else:
            Con_H = C_H*(np.abs(Ve - Volume[index]))/(Vol_OHi + Volume[index])
            pH = -np.log10(Con_H)
            p_H.append(pH)
    display(f"O volume de equivalência é {Ve} mL")        
    return  p_H     
```

```python
y =Calculo_Titulacao(n,Volume)
Calculo_Titulacao(n,Volume)
```

```python
plt.style.use('tableau-colorblind10')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(Volume,y,label = 'adição de H+')
ax.legend(loc = 'lower right')
plt.xlabel('Volume(mL)')
plt.ylabel('pH')
plt.title('Curva de Titulação Base Forte - Ácido Forte')
plt.grid()
plt.ylim(0,14)
plt.xlim(0,10)
fig.tight_layout()
```
